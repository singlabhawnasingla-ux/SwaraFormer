```python
CHUNK_SECONDS = 5
CHUNK_SIZE = 24000 * CHUNK_SECONDS

chunks = []

for start in range(0, len(waveform), CHUNK_SIZE):

    end = min(start + CHUNK_SIZE, len(waveform))

    chunks.append(waveform[start:end])

print("Number of chunks:", len(chunks))
print("First chunk shape:", chunks[0].shape)
```

```python
import torch

embeddings = []

mert.eval()

with torch.no_grad():

    for i, chunk in enumerate(chunks):

        inputs = processor(
            chunk,
            sampling_rate=24000,
            return_tensors="pt"
        )

        inputs = {
            k: v.to(DEVICE)
            for k, v in inputs.items()
        }

        outputs = mert(**inputs)

        emb = outputs.last_hidden_state.squeeze(0).cpu()

        embeddings.append(emb)

        print(f"Chunk {i+1}: {emb.shape}")

        del outputs
        del inputs
        torch.cuda.empty_cache()
```

```python
###############################################################
# MERT EMBEDDING EXTRACTION (RESUMABLE)
###############################################################

import os
import gc
import numpy as np
import pandas as pd
import torch
import torchaudio
from transformers import AutoProcessor, AutoModel

###############################################################
# SETTINGS
###############################################################

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

TARGET_SR = 24000

CHUNK_SECONDS = 30

EMBED_DIR = "embeddings"

os.makedirs(EMBED_DIR, exist_ok=True)

###############################################################
# LOAD MODEL
###############################################################

processor = AutoProcessor.from_pretrained(
    "m-a-p/MERT-v1-330M"
)

mert = AutoModel.from_pretrained(
    "m-a-p/MERT-v1-330M"
)

mert = mert.to(DEVICE)

mert.eval()

print("Running on:", DEVICE)

###############################################################
# EXTRACTION FUNCTION
###############################################################

def extract_song_embedding(audio_path):

    waveform, sr = torchaudio.load(audio_path)

    waveform = waveform.mean(dim=0)

    if sr != TARGET_SR:

        waveform = torchaudio.functional.resample(
            waveform,
            sr,
            TARGET_SR
        )

    chunk_size = CHUNK_SECONDS * TARGET_SR

    song_embeddings = []

    for start in range(0, waveform.shape[0], chunk_size):

        end = min(start + chunk_size, waveform.shape[0])

        chunk = waveform[start:end]

        inputs = processor(

            chunk,

            sampling_rate=TARGET_SR,

            return_tensors="pt"

        )

        inputs = {

            k:v.to(DEVICE)

            for k,v in inputs.items()

        }

        with torch.no_grad():

            outputs = mert(**inputs)

        ########################################################

        # Mean pooling over time

        ########################################################

        emb = outputs.last_hidden_state.mean(dim=1)

        song_embeddings.append(

            emb.squeeze(0).cpu()

        )

        del inputs
        del outputs

        torch.cuda.empty_cache()

    ############################################################

    # Mean pooling over chunks

    ############################################################

    song_embedding = torch.stack(song_embeddings).mean(dim=0)

    return song_embedding.numpy()

###############################################################
# EXTRACTION LOOP
###############################################################

metadata = []

for idx, row in df.iterrows():

    save_path = os.path.join(

        EMBED_DIR,

        f"{idx:04d}.npz"

    )

    ############################################################

    # Skip existing

    ############################################################

    if os.path.exists(save_path):

        print(f"Skipping {idx+1}/{len(df)}")

        metadata.append({

            "file": save_path,

            "title": row.title,

            "raga": row.raga,

            "raga_id": row.raga_id

        })

        continue

    try:

        print("="*60)

        print(f"{idx+1}/{len(df)}")

        print(row.title)

        emb = extract_song_embedding(

            row.audio

        )

        np.savez_compressed(

            save_path,

            embedding=emb,

            raga=row.raga,

            raga_id=row.raga_id,

            title=row.title

        )

        metadata.append({

            "file": save_path,

            "title": row.title,

            "raga": row.raga,

            "raga_id": row.raga_id

        })

        ########################################################

        # Save metadata after EVERY song

        ########################################################

        pd.DataFrame(metadata).to_csv(

            "embeddings.csv",

            index=False

        )

        gc.collect()

        torch.cuda.empty_cache()

    except Exception as e:

        print("FAILED")

        print(e)

        continue

###############################################################
# DONE
###############################################################

print()

print("="*60)

print("Embedding extraction complete.")

print("Embeddings stored in:", EMBED_DIR)y
```

```python
import os
import numpy as np
import pandas as pd
import torch

meta = pd.read_csv("embeddings.csv")

print(meta.head())
```

```python
from torch.utils.data import Dataset

class RagaDataset(Dataset):

    def __init__(self, dataframe):

        self.df = dataframe.reset_index(drop=True)

    def __len__(self):

        return len(self.df)

    def __getitem__(self, idx):

        row = self.df.iloc[idx]

        data = np.load(row.file, allow_pickle=True)

        x = data["embedding"].astype(np.float32)

        y = int(data["raga_id"])

        return torch.tensor(x), torch.tensor(y)
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(

    n_splits=2,

    shuffle=True,

    random_state=42

)
```

```python
import torch.nn as nn

class RagaClassifier(nn.Module):

    def __init__(self, num_classes):

        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(1024,512),

            nn.ReLU(),

            nn.Dropout(0.4),

            nn.Linear(512,256),

            nn.ReLU(),

            nn.Dropout(0.3),

            nn.Linear(256,num_classes)

        )

    def forward(self,x):

        return self.net(x)
```

```python
from torch.utils.data import DataLoader

BATCH_SIZE = 16
```

```python
def train_one_epoch(model,loader,optimizer,criterion):

    model.train()

    loss_sum = 0

    correct = 0

    total = 0

    for x,y in loader:

        x = x.cuda()

        y = y.cuda()

        optimizer.zero_grad()

        pred = model(x)

        loss = criterion(pred,y)

        loss.backward()

        optimizer.step()

        loss_sum += loss.item()

        correct += (pred.argmax(1)==y).sum().item()

        total += len(y)

    return loss_sum/len(loader),correct/total
```

```python
def evaluate(model,loader):

    model.eval()

    preds=[]

    labels=[]

    with torch.no_grad():

        for x,y in loader:

            x=x.cuda()

            pred=model(x)

            preds.extend(pred.argmax(1).cpu().numpy())

            labels.extend(y.numpy())

    return labels,preds
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
for fold,(train_idx,test_idx) in enumerate(

    skf.split(meta,meta.raga_id)

):

    print("="*60)

    print("Fold",fold+1)

    train_df=meta.iloc[train_idx]

    test_df=meta.iloc[test_idx]

    train_loader=DataLoader(

        RagaDataset(train_df),

        batch_size=16,

        shuffle=True

    )

    test_loader=DataLoader(

        RagaDataset(test_df),

        batch_size=16

    )

    model=RagaClassifier(

        meta.raga_id.nunique()

    ).cuda()

    optimizer=torch.optim.Adam(

        model.parameters(),

        lr=1e-4

    )

    criterion=nn.CrossEntropyLoss()

    for epoch in range(20):

        loss,acc=train_one_epoch(

            model,

            train_loader,

            optimizer,

            criterion

        )

        print(epoch+1,loss,acc)

    labels,preds=evaluate(

        model,

        test_loader

    )

    print()

    print("Accuracy",

          accuracy_score(labels,preds))

    print("Precision",

          precision_score(labels,preds,average="weighted"))

    print("Recall",

          recall_score(labels,preds,average="weighted"))

    print("F1",

          f1_score(labels,preds,average="weighted"))
```
